### Numerical Analysis

```python
numerical_col = ['Estimated Income','Superannuation Savings','Credit Card Balance','Bank Loans',
                 'Bank Deposits','Business Lending','Saving Accounts','Checking Accounts']
#univariate analysis and visualization
plt.figure(figsize=(12,16))
for i,col in enumerate(numerical_col):
    plt.subplot(4,2, i+1)
    sns.histplot(df[col],kde=True)
    plt.title(col)
    plt.tight_layout()
    
plt.show()
```

#### heatmaps

```python
numerical_col = ['Estimated Income','Superannuation Savings','Credit Card Balance','Bank Loans',
                 'Bank Deposits','Business Lending','Saving Accounts','Checking Accounts','Foreign Currency Account']
correlation_matrix = df[numerical_col].corr()
plt.figure(figsize=(12,12))
sns.heatmap(correlation_matrix,annot=True,cmap='crest',fmt='.2f')
```
